# Cleaning Data

**Membersihkan data customers_df**

*   Menghilangkan duplicate data



```python
# Menghapus duplikasi pada customers_df
customers_df.drop_duplicates(inplace=True)
```

```python
# Memeriksa kembali apakah terdapat duplikasi terhadap data customers_df
print("Jumlah duplikasi: ", customers_df.duplicated().sum())
```

*   Menangani Missing Value



```python
# Melihat data missing value pada customers_df
# customers_df[customers_df.isna()]
customers_df[customers_df.isna().any(axis=1)]
```

```python
# Mengidentifikasi nilai yang dominan pada kolom gender
customers_df.gender.value_counts()
```

```python
# Mengganti nilai NaN menjadi "Prefer not to say" pada kolom gender
customers_df.fillna(value="Prefer not to say", inplace=True)
```

```python
# Menjalankan kembali kode mengidentifikasi missing value
customers_df.isna().sum()
```

*   Menangani inaccurate value


```python
# Teknik filtering melihat data yang mengandung inaccurate value
customers_df[customers_df.age == customers_df.age.max()]
```

```python
# Mengganti nilai age dari 700 menjadi 70
customers_df.age.replace(customers_df.age.max(), 70, inplace=True)
```

```python
# Mengecek kembali data yang mengandung inaccurate value
customers_df[customers_df.age == customers_df.age.max()]
```

```python
# Mengganti nilai age dari 500 menjadi 50
customers_df.age.replace(customers_df.age.max(), 50, inplace=True)
```

```python
customers_df.describe()
```

**Membersihkan data orders_df**

```python
# Merubah tipe data pada kolom order_date & delivery_date menjadi datetime
datetime_columns = ["order_date", "delivery_date"]
for column in datetime_columns:
  orders_df[column] = pd.to_datetime(orders_df[column])
```

```python
orders_df.info()
```

**Membersihkan data product_df**

```python
# Membuang data yang duplikat pada product_df menggunakan drop_duplicates()
product_df.drop_duplicates(inplace=True)
```

```python
# Memastikan dan Menampilkan jumlah data yang duplikat
print("Jumlah duplikasi: ", product_df.duplicated().sum())
```

**Membersihkan data sales_df**

```python
# Melihat baris data yang mengandung missing value pada sales_df di kolom total_price
# sales_df[sales_df["total_price"].isna()]
sales_df[sales_df.total_price.isna()]
```

```python
# Mengisi kolom total_price dengan perkalian antara price_per_unit dan quantity
# sales_df["total_price"].fillna(sales_df["price_per_unit"] * sales_df["quantity"], inplace=True)
sales_df["total_price"] = sales_df["price_per_unit"] * sales_df["quantity"]
```

```python
sales_df.isna().sum()
```
